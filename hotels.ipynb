```python
resort_len= len(df[df['hotel']=='Resort Hotel'])
city_len=len(df[df['hotel']=='City Hotel'])

hotel_types = ['Resort Hotel', 'City Hotel']
lengths = [resort_len, city_len]

plt.bar(hotel_types, lengths)
plt.xlabel('Hotel Type')
plt.ylabel('Number of Guest')
plt.title('Number of Guest for Each Hotel Type')
plt.show()
```

```python
country_wise_counts = df[df['is_canceled'] == 0]['country'].value_counts().reset_index()
country_wise_counts.columns = ['country', 'No of guests']
country_wise_counts
```

```python
import folium
from folium.plugins import HeatMap
import plotly.express as px
```

```python
basemap = folium.Map()
guests_map = px.choropleth(country_wise_counts, locations = country_wise_counts['country'],
                           color = country_wise_counts['No of guests'], hover_name = country_wise_counts['country'])
guests_map.show()
```

```python
data = df[df['is_canceled'] == 0]
px.box(data_frame = data, x = 'reserved_room_type', y = 'adr', color = 'hotel', template = 'plotly_dark')
```

```python
data_resort = data[(data['hotel'] == 'Resort Hotel') & (data['is_canceled'] == 0)]
data_city = data[(data['hotel'] == 'City Hotel') & (data['is_canceled'] == 0)]
```

```python
resort_hotel = data_resort.groupby(['arrival_date_month'])['adr'].mean().reset_index()
resort_hotel
```

```python
city_hotel = data_city.groupby(['arrival_date_month'])['adr'].mean().reset_index()
city_hotel
```
